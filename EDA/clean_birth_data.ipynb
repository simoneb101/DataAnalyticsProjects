```python
import pandas as pd
import numpy as np
from collections import Counter

layout_path = r"C:\Users\sboyd\OneDrive\Desktop\Data Analyst Projects\Birth Analysis\data\UserGuide2024_layout_pages9_38.xlsx"
path = r"C:\Users\sboyd\OneDrive\Desktop\Data Analyst Projects\Birth Analysis\data\Nat2024PublicUS.c20250512.r20250708.txt"
output_path = r"C:\Users\sboyd\OneDrive\Desktop\Data Analyst Projects\Birth Analysis\data\birth.csv"

layout = pd.read_excel(layout_path, sheet_name="Layout")

#Widths
col_widths = layout["Length"].astype(int).tolist()

#Raw field names
raw_names = layout["Field"].astype(str).str.strip().tolist()

#Make all names unique
counts = Counter()
col_names = []
for name in raw_names:
    counts[name] += 1
    suffix = counts[name]
    col_names.append(name if suffix == 1 else f"{name}_{suffix}")

#Read as strings first
dtype_map = {name: "string" for name in col_names}

df = pd.read_fwf(
    path,
    widths=col_widths,
    names=col_names,
    dtype=dtype_map,
)

#Optional: remove filler columns
df = df.loc[:, ~df.columns.str.startswith("FILLER")]

print(df.head())
print(df.columns.tolist())

#Output file
df.to_csv(output_path, index=False, header=True)
```
