```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from pathlib import Path
import os
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
from langchain.agents import create_agent
from langchain_core.messages import HumanMessage

API_KEY = os.getenv("OPEN_AI")

f = Path.cwd().__str__() + "\\data_sources\\HR data\\Messy_Employee_dataset.csv"

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

hr_df = pd.read_csv(f)
print(hr_df.head())

llm = ChatOpenAI(
    model="gpt-5-nano",
    # stream_usage=True,
    # temperature=None,
    # max_tokens=None,
    # timeout=None,
    # reasoning_effort="low",
    # max_retries=2,
    api_key=API_KEY,  # If you prefer to pass api key in directly
    # base_url="...",
    # organization="...",
    # other params...
)

@tool
def analyze_df(code: str) -> str:
    """Run pandas code on df: df.head(), df.describe(), df.groupby(...)"""
    env = {"df": hr_df, "pd": pd}
    exec(code, env)
    return str(env.get("result", "Set result = ..."))

llm = ChatOpenAI(model="gpt-5.1", api_key=API_KEY)
agent = create_agent(llm, [analyze_df])

response = agent.invoke({
    "messages": [HumanMessage(content="Generate the necessary data cleansing techniques needed on this dataframe, break it out in a dictionary format.")]
})

print(response)
```
