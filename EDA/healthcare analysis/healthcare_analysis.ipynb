```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from pathlib import Path

f = Path.cwd().__str__() + "\\data_sources\\healthcare data\\health_kaggle_data_realistic.csv"

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

hlth_df = pd.read_csv(f)

#Clean Patient Name
#hlth_df['First Name'] = hlth_df.apply(lambda x: x['Name'].split(' ')[0].capitalize(), axis=1)
#hlth_df['Last Name'] = hlth_df.apply(lambda x: x['Name'].split(' ')[1].capitalize(), axis=1)

#Drop Name
#hlth_df.drop(['Name'], axis=1, inplace=True)

#Fix the date columns into date type
hlth_df['Date of Admission'] = pd.to_datetime(hlth_df['Date of Admission'])
hlth_df['Discharge Date'] = pd.to_datetime(hlth_df['Discharge Date'])

#Calculate admission_duration_days
hlth_df['Admission Duration Days'] = (hlth_df['Discharge Date'] - hlth_df['Date of Admission']).dt.days


hlth_df['Hospital with Room Number'] = hlth_df.apply(lambda x: f"{x['Hospital']}-RM#{x['Room Number']}", axis=1)
hlth_df.head().style.hide(axis='index')
#hlth_df.columns
```

```python
##Summarize Billing information by Hospital and Doctor
billSummary = hlth_df.groupby(['Hospital', 'Doctor'])['Billing Amount'].sum().rename('Total Billing Amount').reset_index()

billSummary.sort_values(by='Total Billing Amount', ascending=False).head().style.hide(axis='index')
```

```python
data = hlth_df.groupby('Blood Type')['Blood Type'].size().rename('Total Blood Type').reset_index()

fig = px.bar(data, x='Blood Type', y='Total Blood Type', title="Total by Blood Type")

fig.show()
```

```python
hlth_df['Admission Year'] = hlth_df['Date of Admission'].dt.year

data = hlth_df.groupby('Admission Year').size().rename('Total').reset_index()
fig = px.line(data, x="Admission Year", y="Total", title="Total by Year of Admission")

fig.show()
```

```python
data = hlth_df.groupby('Blood Type')['Admission Duration Days'].sum().rename('Total Admission Duration in Days').reset_index()

fig = px.bar(data, x='Blood Type', y='Total Admission Duration in Days', title="Total Admission Duration in Days by Blood Type")

fig.show()
```

```python
data = hlth_df.groupby('Age')['Admission Duration Days'].mean().rename('Average Admission Duration in Days').reset_index()

fig = px.bar(data, x='Age', y='Average Admission Duration in Days', title="Average Admission Duration in Days by Age")

fig.show()
```
